Ler um valor para a variável X, multiplique este valor por 12, implicando-o à variável de resposta R e apresente o valor de R obtido, repetindo esta sequência por quatro vezes.

```python
"""
Pseudocódigo

Programa multiplicaDoze
var
    x, r : real
    cont : int
Inicio
    escreva("Digite um valor: ")
    leia(x)
    para cont de 1 até 4 passo 1 faça
      r = <- * 12
      escreva("Resultado = " + r)
"""
```

```python
valor = float(input("Digite um valor: "))

for x in range(4):
  resultado = valor * 12
  valor = resultado
  print(f"Resultado {x + 1} = {resultado:.2f}")
```
